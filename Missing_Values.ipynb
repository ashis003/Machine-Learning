```python
import pandas as pd
```

```python
myDatasets = {
    'Name' : ['Mamma','Ashis','Surya','Bisso','Buchi','Ishmita'],
    'Roll_no' : [17,16,1,3,7,-1],
    'Fab_Subjects' : ['Hindi','not available','English','Science','Sanskrit','Arts'],
    'Marks' : [90,95,89,-1,88,92]
}
df=pd.DataFrame(myDatasets)
df
```

```python
df.to_csv('newDatasets.csv')
```

```python
df=pd.read_csv('newDatasets.csv',na_values=['not available',-1])
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

```python
newdf=df.fillna(0)
newdf
```

```python
newdf=df.fillna({
   'Roll_no' : 5,
   'Fab_Subjects' : 'Math',
   'Marks' : 87

})
newdf
```

```python
newdf=df.fillna(method='ffill')
newdf
```

```python
newdf = df.interpolate()
newdf
```

newdf=df.dropna()
newdf

```python
newdf=df.dropna(how='all')
newdf
```
